```python
# điều chỉnh dữ liệu để dố liệu đúng với logic 
# Cập nhật giá trị null cho SHORT_TERM_COUNT
df1['SHORT_TERM_COUNT'] = np.where(
    df1['SHORT_TERM_COUNT'].isnull() & df1['SHORT_TERM_COUNT_BANK'].notnull() & df1['SHORT_TERM_COUNT_NON_BANK'].notnull(),
    df1['SHORT_TERM_COUNT_BANK'] + df1['SHORT_TERM_COUNT_NON_BANK'],
    df1['SHORT_TERM_COUNT']
)

# Cập nhật giá trị null cho SHORT_TERM_COUNT_BANK
df1['SHORT_TERM_COUNT_BANK'] = np.where(
    df1['SHORT_TERM_COUNT_BANK'].isnull() & df1['SHORT_TERM_COUNT'].notnull() & df1['SHORT_TERM_COUNT_NON_BANK'].notnull(),
    df1['SHORT_TERM_COUNT'] + 1 - df1['SHORT_TERM_COUNT_NON_BANK'],
    df1['SHORT_TERM_COUNT_BANK']
)

# Cập nhật giá trị null cho SHORT_TERM_COUNT_NON_BANK
df1['SHORT_TERM_COUNT_NON_BANK'] = np.where(
    df1['SHORT_TERM_COUNT_NON_BANK'].isnull() & df1['SHORT_TERM_COUNT'].notnull() & df1['SHORT_TERM_COUNT_BANK'].notnull(),
    df1['SHORT_TERM_COUNT'] + 1 - df1['SHORT_TERM_COUNT_BANK'],
    df1['SHORT_TERM_COUNT_NON_BANK']
)

# Tương tự cho MID_TERM_COUNT và các cột liên quan
df1['MID_TERM_COUNT'] = np.where(
    df1['MID_TERM_COUNT'].isnull() & df1['MID_TERM_COUNT_BANK'].notnull() & df1['MID_TERM_COUNT_NON_BANK'].notnull(),
    df1['MID_TERM_COUNT_BANK'] + df1['MID_TERM_COUNT_NON_BANK'] ,
    df1['MID_TERM_COUNT']
)

df1['MID_TERM_COUNT_BANK'] = np.where(
    df1['MID_TERM_COUNT_BANK'].isnull() & df1['MID_TERM_COUNT'].notnull() & df1['MID_TERM_COUNT_NON_BANK'].notnull(),
    df1['MID_TERM_COUNT'] + 1 - df1['MID_TERM_COUNT_NON_BANK'],
    df1['MID_TERM_COUNT_BANK']
)

df1['MID_TERM_COUNT_NON_BANK'] = np.where(
    df1['MID_TERM_COUNT_NON_BANK'].isnull() & df1['MID_TERM_COUNT'].notnull() & df1['MID_TERM_COUNT_BANK'].notnull(),
    df1['MID_TERM_COUNT'] + 1 - df1['MID_TERM_COUNT_BANK'],
    df1['MID_TERM_COUNT_NON_BANK']
)

# Tương tự cho LONG_TERM_COUNT và các cột liên quan
df1['LONG_TERM_COUNT'] = np.where(
    df1['LONG_TERM_COUNT'].isnull() & df1['LONG_TERM_COUNT_BANK'].notnull() & df1['LONG_TERM_COUNT_NON_BANK'].notnull(),
    df1['LONG_TERM_COUNT_BANK'] + df1['LONG_TERM_COUNT_NON_BANK'] ,
    df1['LONG_TERM_COUNT']
)

df1['LONG_TERM_COUNT_BANK'] = np.where(
    df1['LONG_TERM_COUNT_BANK'].isnull() & df1['LONG_TERM_COUNT'].notnull() & df1['LONG_TERM_COUNT_NON_BANK'].notnull(),
    df1['LONG_TERM_COUNT'] + 1 - df1['LONG_TERM_COUNT_NON_BANK'],
    df1['LONG_TERM_COUNT_BANK']
)

df1['LONG_TERM_COUNT_NON_BANK'] = np.where(
    df1['LONG_TERM_COUNT_NON_BANK'].isnull() & df1['LONG_TERM_COUNT'].notnull() & df1['LONG_TERM_COUNT_BANK'].notnull(),
    df1['LONG_TERM_COUNT'] + 1 - df1['LONG_TERM_COUNT_BANK'],
    df1['LONG_TERM_COUNT_NON_BANK']
)

# Kiểm tra lại DataFrame để xem các giá trị null đã được điền
print(df1.isnull().sum())
```

#### NUMBER_OF_LOANS 

```python
# Tạo cột NUMBER_OF_LOANS_1 = NUMBER_OF_LOANS + 1
df1['NUMBER_OF_LOANS_1'] = df1['NUMBER_OF_LOANS'] + 1

# Điền giá trị cho NUMBER_OF_LOANS_1 khi BANK và NON_BANK không null
df1['NUMBER_OF_LOANS_1'] = np.where(
    df1['NUMBER_OF_LOANS_1'].isnull() & df1['NUMBER_OF_LOANS_BANK'].notnull() & df1['NUMBER_OF_LOANS_NON_BANK'].notnull(),
    df1['NUMBER_OF_LOANS_BANK'] + df1['NUMBER_OF_LOANS_NON_BANK'],
    df1['NUMBER_OF_LOANS_1']
)

# Điền giá trị cho NUMBER_OF_LOANS_BANK khi NUMBER_OF_LOANS_1 và NON_BANK không null
df1['NUMBER_OF_LOANS_BANK'] = np.where(
    df1['NUMBER_OF_LOANS_BANK'].isnull() & df1['NUMBER_OF_LOANS_1'].notnull() & df1['NUMBER_OF_LOANS_NON_BANK'].notnull(),
    df1['NUMBER_OF_LOANS_1'] - df1['NUMBER_OF_LOANS_NON_BANK'],
    df1['NUMBER_OF_LOANS_BANK']
)

# Điền giá trị cho NUMBER_OF_LOANS_NON_BANK khi NUMBER_OF_LOANS_1 và BANK không null
df1['NUMBER_OF_LOANS_NON_BANK'] = np.where(
    df1['NUMBER_OF_LOANS_NON_BANK'].isnull() & df1['NUMBER_OF_LOANS_1'].notnull() & df1['NUMBER_OF_LOANS_BANK'].notnull(),
    df1['NUMBER_OF_LOANS_1'] - df1['NUMBER_OF_LOANS_BANK'],
    df1['NUMBER_OF_LOANS_NON_BANK']
)

```

```python
# Tách df1 thành hai DataFrame dựa vào cột label
df_label_0 = df1[df1['label'] == 0].copy()
df_label_1 = df1[df1['label'] == 1].copy()

# Điền các giá trị null trong df_label_0
mode_bank_0 = df_label_0['NUMBER_OF_LOANS_BANK'].mode()[0] if not df_label_0['NUMBER_OF_LOANS_BANK'].mode().empty else np.nan
mode_non_bank_0 = df_label_0['NUMBER_OF_LOANS_NON_BANK'].mode()[0] if not df_label_0['NUMBER_OF_LOANS_NON_BANK'].mode().empty else np.nan

df_label_0['NUMBER_OF_LOANS_BANK'].fillna(mode_bank_0, inplace=True)
df_label_0['NUMBER_OF_LOANS_NON_BANK'].fillna(mode_non_bank_0, inplace=True)

# Điền các giá trị null trong df_label_1
mode_bank_1 = df_label_1['NUMBER_OF_LOANS_BANK'].mode()[0] if not df_label_1['NUMBER_OF_LOANS_BANK'].mode().empty else np.nan
mode_non_bank_1 = df_label_1['NUMBER_OF_LOANS_NON_BANK'].mode()[0] if not df_label_1['NUMBER_OF_LOANS_NON_BANK'].mode().empty else np.nan

df_label_1['NUMBER_OF_LOANS_BANK'].fillna(mode_bank_1, inplace=True)
df_label_1['NUMBER_OF_LOANS_NON_BANK'].fillna(mode_non_bank_1, inplace=True)

# Gộp lại hai DataFrame sau khi điền giá trị null vào thành df1
df1_filled = pd.concat([df_label_0, df_label_1])

# Tính toán lại các giá trị của NUMBER_OF_LOANS_1 dựa vào NUMBER_OF_LOANS_BANK và NUMBER_OF_LOANS_NON_BANK
df1_filled['NUMBER_OF_LOANS_1'] = df1_filled['NUMBER_OF_LOANS_BANK'] + df1_filled['NUMBER_OF_LOANS_NON_BANK']

# Hiển thị kết quả sau khi điền giá trị null và tính toán
df1_filled
```

```python
# Kiểm tra xem có bao nhiêu dòng thỏa mãn điều kiện
matching_rows_cc = df4[(df4['ENQUIRIES_FOR_CC_12M'] + 2) == 
                        (df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M'] + df4['ENQUIRIES_FROM_BANK_FOR_CC_12M'])]

# In ra số lượng dòng thỏa mãn điều kiện
print(f"Số dòng có ENQUIRIES_FOR_CC_12M + 2 = ENQUIRIES_FROM_NON_BANK_FOR_CC_12M + ENQUIRIES_FROM_BANK_FOR_CC_12M: {len(matching_rows_cc)}")

# Tính toán lại các giá trị cho cột 'ENQUIRIES_FOR_CC_12M' (chỉ tính +2 ở các giá trị không null, null thì giữ nguyên)
df4['ENQUIRIES_FOR_CC_12M'] = df4['ENQUIRIES_FOR_CC_12M'].apply(lambda x: x + 2 if pd.notnull(x) else x)

# Điền giá trị null trong cột 'ENQUIRIES_FOR_CC_12M'
df4.loc[df4['ENQUIRIES_FOR_CC_12M'].isnull(), 'ENQUIRIES_FOR_CC_12M'] = (
    df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M'] + 
    df4['ENQUIRIES_FROM_BANK_FOR_CC_12M']
)

# Tính toán giá trị null cho cột ENQUIRIES_FROM_NON_BANK_FOR_CC_12M
df4.loc[df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M'].isnull(), 'ENQUIRIES_FROM_NON_BANK_FOR_CC_12M'] = (
    df4['ENQUIRIES_FOR_CC_12M'] - df4['ENQUIRIES_FROM_BANK_FOR_CC_12M']
)

# Tính toán giá trị null cho cột ENQUIRIES_FROM_BANK_FOR_CC_12M
df4.loc[df4['ENQUIRIES_FROM_BANK_FOR_CC_12M'].isnull(), 'ENQUIRIES_FROM_BANK_FOR_CC_12M'] = (
    df4['ENQUIRIES_FOR_CC_12M'] - df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M']
)
```

```python
# Tính toán giá trị chênh lệch
df4['x'] = (
    df4['ENQUIRIES_FROM_NON_BANK_FOR_LOAN_12M'] + 
    df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M'] - 
    df4['ENQUIRIES_FROM_NON_BANK_12M']
)

# Giữ nguyên giá trị null nếu một trong ba cột có giá trị null
df4['x'] = df4['x'].where(
    df4['ENQUIRIES_FROM_NON_BANK_FOR_LOAN_12M'].notnull() & 
    df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M'].notnull() & 
    df4['ENQUIRIES_FROM_NON_BANK_12M'].notnull()
)

# Đếm số lượng các giá trị khác nhau trong cột x
count_x = df4.groupby('x').size()

# In ra kết quả
print("Số lượng giá trị chênh lệch (x) trong cột:")
print(count_x)

# Tính toán lại các giá trị cho cột ENQUIRIES_FROM_NON_BANK_12M
df4['ENQUIRIES_FROM_NON_BANK_12M'] = df4['ENQUIRIES_FROM_NON_BANK_12M'].apply(lambda x: x + 2 if pd.notnull(x) else x)

# Tính toán và điền giá trị null trong cột 'ENQUIRIES_FROM_NON_BANK_12M'
calculated_values = df4['ENQUIRIES_FROM_NON_BANK_FOR_LOAN_12M'] + df4['ENQUIRIES_FROM_NON_BANK_FOR_CC_12M']

# Cập nhật các giá trị null trong cột ENQUIRIES_FROM_NON_BANK_12M
df4['ENQUIRIES_FROM_NON_BANK_12M'] = df4['ENQUIRIES_FROM_NON_BANK_12M'].fillna(calculated_values)

# So sánh và cập nhật các giá trị không null nếu không giống với giá trị tính toán
df4.loc[
    df4['ENQUIRIES_FROM_NON_BANK_12M'].notnull() & 
    (df4['ENQUIRIES_FROM_NON_BANK_12M'] != calculated_values), 
    'ENQUIRIES_FROM_NON_BANK_12M'
] = calculated_values    
```

```python
# Tính toán giá trị chênh lệch
df4['x'] = (
    df4['ENQUIRIES_FROM_BANK_FOR_LOAN_12M'] + 
    df4['ENQUIRIES_FROM_BANK_FOR_CC_12M'] - 
    df4['ENQUIRIES_FROM_BANK_12M']
)

# Giữ nguyên giá trị null nếu một trong ba cột có giá trị null
df4['x'] = df4['x'].where(
    df4['ENQUIRIES_FROM_BANK_FOR_LOAN_12M'].notnull() & 
    df4['ENQUIRIES_FROM_BANK_FOR_CC_12M'].notnull() & 
    df4['ENQUIRIES_FROM_BANK_12M'].notnull()
)

# Đếm số lượng các giá trị khác nhau trong cột x
count_x = df4.groupby('x').size()

# In ra kết quả
print("Số lượng giá trị chênh lệch (x) trong cột:")
print(count_x)

# Tính toán lại các giá trị cho cột ENQUIRIES_FROM_BANK_12M
df4['ENQUIRIES_FROM_BANK_12M'] = df4['ENQUIRIES_FROM_BANK_12M'].apply(lambda x: x + 2 if pd.notnull(x) else x)

# Tính toán và điền giá trị null trong cột 'ENQUIRIES_FROM_BANK_12M'
calculated_values = df4['ENQUIRIES_FROM_BANK_FOR_LOAN_12M'] + df4['ENQUIRIES_FROM_BANK_FOR_CC_12M']

# Cập nhật các giá trị null trong cột ENQUIRIES_FROM_BANK_12M
df4['ENQUIRIES_FROM_BANK_12M'] = df4['ENQUIRIES_FROM_BANK_12M'].fillna(calculated_values)

# So sánh và cập nhật các giá trị không null nếu không giống với giá trị tính toán
df4.loc[
    df4['ENQUIRIES_FROM_BANK_12M'].notnull() & 
    (df4['ENQUIRIES_FROM_BANK_12M'] != calculated_values), 
    'ENQUIRIES_FROM_BANK_12M'
] = calculated_values

# Tạo cột trung gian để tính toán `ENQUIRIES_12M + 6`
df4['ENQUIRIES_12M_adjusted'] = df4['ENQUIRIES_12M'] + 6

# Duyệt qua từng dòng và điền giá trị vào các cột bị thiếu
for index, row in df4.iterrows():
    # Nếu `ENQUIRIES_FROM_BANK_12M` bị null nhưng `ENQUIRIES_FROM_NON_BANK_12M` không null
    if pd.isnull(row['ENQUIRIES_FROM_BANK_12M']) and not pd.isnull(row['ENQUIRIES_FROM_NON_BANK_12M']):
        # Điền giá trị còn thiếu dựa vào ENQUIRIES_12M_adjusted - ENQUIRIES_FROM_NON_BANK_12M
        df4.loc[index, 'ENQUIRIES_FROM_BANK_12M'] = row['ENQUIRIES_12M_adjusted'] - row['ENQUIRIES_FROM_NON_BANK_12M']
    
    # Nếu `ENQUIRIES_FROM_NON_BANK_12M` bị null nhưng `ENQUIRIES_FROM_BANK_12M` không null
    elif pd.isnull(row['ENQUIRIES_FROM_NON_BANK_12M']) and not pd.isnull(row['ENQUIRIES_FROM_BANK_12M']):
        # Điền giá trị còn thiếu dựa vào ENQUIRIES_12M_adjusted - ENQUIRIES_FROM_BANK_12M
        df4.loc[index, 'ENQUIRIES_FROM_NON_BANK_12M'] = row['ENQUIRIES_12M_adjusted'] - row['ENQUIRIES_FROM_BANK_12M']

# Xóa cột trung gian `ENQUIRIES_12M_adjusted`
df4.drop(columns=['ENQUIRIES_12M_adjusted'], inplace=True)

# Tính toán lại giá trị cho cột `ENQUIRIES_12M`
df4['ENQUIRIES_12M'] = df4['ENQUIRIES_FROM_BANK_12M'] + df4['ENQUIRIES_FROM_NON_BANK_12M']

# Kiểm tra điều kiện cột `ENQUIRIES_12M` là null
null_mask = df4['ENQUIRIES_12M'].isnull()

# Tính tổng `ENQUIRIES_FOR_LOAN_12M` và `ENQUIRIES_FOR_CC_12M` cho các dòng có `ENQUIRIES_12M` là null
df4.loc[null_mask, 'ENQUIRIES_12M'] = df4.loc[null_mask, 'ENQUIRIES_FOR_LOAN_12M'] + df4.loc[null_mask, 'ENQUIRIES_FOR_CC_12M']

# Kiểm tra lại các giá trị null sau khi đã điền
print(f"Số giá trị null còn lại trong cột ENQUIRIES_12M: {df4['ENQUIRIES_12M'].isnull().sum()}")
```
